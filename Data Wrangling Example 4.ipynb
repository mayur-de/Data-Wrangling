## Import Libraries and Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as datetime
```

```python
data = pd.read_excel("Badly-Structured-Sales-Data-4.xlsx")
data
```

## **Task:** Transform the data so that it can be used for analysis. 

### **Proposed Solution:** 
- Create 5 columns
    - Order Date
    - Segment
    - Order ID
    - Ship Mode
    - Sales

- **Workflow:**
- `Drop irrelevant rows:` Delete columns with `Total` in it. (`Grand Total`)
- `Handle Column Headers:`
    - Rename columns with Unnamed with the previous segment name, as it is logical to assume that these are the headers under segment mentioned in previous column.
- Handle Columns that has information on `Order Date` and `Order ID`
- Merge to get long formatted data

```python
# `Drop irrelevant rows:` Delete row with `Total` in it. (`Grand Total`)

data = data[data['Ship Mode'] != 'Grand Total']
data
```

```python
# Fill Unnamed column headers with previous headers

new_columns = []

for i in range(len(data.columns)):
    if "Unnamed" in data.columns[i] and data.columns[i] != "Unnamed: 0":
        new_columns.append(new_columns[i - 1])
    else:
        new_columns.append(data.columns[i])

new_columns
```

```python
# Assign new_column to data.column
data.columns = new_columns

# Reset index
data.reset_index(drop=True, inplace= True)
data
```

```python
# Merge column header + first row to get ship mode_segment

for i, j in enumerate(data.columns):
    data.columns.values[i] = f"{j}_{data.iloc[0, i]}"
    
data.head()
```

```python
# drop index= 0 and 1 in axis = 0
data.drop(index=[0, 1], axis= 0, inplace=True)

# Reset index
data.reset_index(drop=True, inplace= True)
data.head()
```

```python
# Rename column "Unnamed: 0_nan" and "Ship Mode_Segment" with Order ID and Order Date for calarity

data.rename(columns= {"Unnamed: 0_nan": "Order ID",
                      "Ship Mode_Segment": "Order Date"}, inplace=True)
data.head()
```

```python
# MElt data

data_melted = data.melt(id_vars= ['Order ID', 'Order Date'],
                        var_name= "ShipMode_Segment",
                        value_name= "Sales")

data_melted
```

```python
# Split "ShipMode_Segment"

data_melted[['Ship Mode', 'Segment']] = data_melted['ShipMode_Segment'].str.split("_", expand= True)

# Drop column "ShipMode_Segment"
data_melted.drop(columns= "ShipMode_Segment", inplace= True)
data_melted
```

```python
# Filter Sales to remove NaN

data_melted = data_melted[data_melted['Sales'].notnull()]
data_melted
```
